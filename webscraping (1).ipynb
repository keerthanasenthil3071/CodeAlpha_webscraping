# 🌐 Web Scraping – Extracting Product Data from a Website

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://quotes.toscrape.com/"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
quotes = soup.find_all("span", class_="text")
authors = soup.find_all("small", class_="author")

quote_list = []
author_list = []

for i in range(len(quotes)):
    quote_list.append(quotes[i].text)
    author_list.append(authors[i].text)
```

```python
data = pd.DataFrame({
    "Quote": quote_list,
    "Author": author_list
})

data.head()
```

```python
data.to_csv("quotes_data.csv", index=False)
```

```python
import matplotlib.pyplot as plt

data['Author'].value_counts().head().plot(kind='bar')
plt.title("Top Authors")
plt.show()
```

## Observations

- Quotes are collected from the website successfully.
- Different authors have contributed multiple quotes.
- Some authors appear more frequently.
- Web scraping helps in collecting textual data easily.
